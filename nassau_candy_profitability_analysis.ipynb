# Product Line Profitability & Margin Performance Analysis
### Nassau Candy Distributor
This notebook includes data cleaning, KPI calculations, profitability analysis, and visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df = pd.read_csv('Nassau Candy Distributor.csv')
df.head()
```

## Data Cleaning

```python
df = df[df['Sales'] > 0]
df = df[df['Gross Profit'] >= 0]
df['Units'] = df['Units'].fillna(1)
```

## Profitability Metrics

```python
df['Gross Margin %'] = (df['Gross Profit'] / df['Sales']) * 100
df['Profit per Unit'] = df['Gross Profit'] / df['Units']
df['Revenue Contribution'] = df['Sales'] / df['Sales'].sum()
df['Profit Contribution'] = df['Gross Profit'] / df['Gross Profit'].sum()
df.head()
```

## Product Level Profitability

```python
product_profit = df.groupby('Product Name').agg({'Sales':'sum','Gross Profit':'sum','Units':'sum','Gross Margin %':'mean'}).sort_values(by='Gross Profit', ascending=False)
product_profit.head(10)
```

## Division Analysis

```python
division_analysis = df.groupby('Division').agg({'Sales':'sum','Gross Profit':'sum','Gross Margin %':'mean'}).sort_values(by='Gross Profit', ascending=False)
division_analysis
```

## Cost vs Sales Scatter Plot

```python
plt.figure()
sns.scatterplot(data=df, x='Cost', y='Sales', hue='Division')
plt.title('Cost vs Sales Analysis')
plt.show()
```

## Top 10 Profitable Products

```python
top_products = df.groupby('Product Name')['Gross Profit'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar')
plt.title('Top 10 Profitable Products')
plt.ylabel('Gross Profit')
plt.show()
```

## Pareto Analysis

```python
pareto = df.groupby('Product Name')['Gross Profit'].sum().sort_values(ascending=False)
pareto_cumsum = pareto.cumsum() / pareto.sum()
plt.figure()
pareto.plot(kind='bar')
plt.title('Pareto Profit Distribution')
plt.show()
```

## Save Clean Dataset

```python
df.to_csv('clean_nassau_data.csv', index=False)
print('Clean dataset saved')
```
